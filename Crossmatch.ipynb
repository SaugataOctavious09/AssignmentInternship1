```python
import pandas as pd
import numpy as np
```

```python
Amazon = pd.read_excel("amz_com-ecommerce_sample_1.xlsx")
Amazon = Amazon[["product_name","retail_price","discounted_price"]]
Amazon = Amazon.rename(columns={"product_name":"Amazon_product_name","retail_price":"amazon_retail_price","discounted_price":"Amazon_discounted_price"})
Amazon["Product"] = Amazon["Amazon_product_name"]
```

```python
Amazon
```

```python
Flipkart = pd.read_excel("flipkart_com-ecommerce_sample_1.xlsx")
Flipkart = Flipkart[["product_name","retail_price","discounted_price"]]
Flipkart = Flipkart.rename(columns={"product_name":"Flipkart_product_name","retail_price":"flipkart_retail_price","discounted_price":"flipkart_discounted_price"})
Flipkart["Product"] = Flipkart["Flipkart_product_name"]
```

```python
Flipkart
```

```python
df_search = pd.merge(Flipkart, Amazon, on = 'Product')
df_search
```

```python
df_search = df_search.drop_duplicates()
df_search = df_search.reset_index()
```
